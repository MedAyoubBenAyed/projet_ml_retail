Appliquation de la mathode VIF 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

Affichages VIF : état initial, matrice de corrélation, suppression itérative et variables finales retenues

```python
# Séparer X et y pour le VIF
X = df_reduced.drop(columns=["Churn"], errors="ignore").copy()

# Garder seulement les variables numériques
X_numeric = X.select_dtypes(include=[np.number]).copy()

# Nettoyage minimal pour rendre le VIF calculable
X_numeric.replace([np.inf, -np.inf], np.nan, inplace=True)
X_numeric.fillna(X_numeric.median(numeric_only=True), inplace=True)

# Retirer les colonnes constantes ou quasi-constantes
const_cols = X_numeric.nunique(dropna=False)[X_numeric.nunique(dropna=False) <= 1].index.tolist()
if const_cols:
    X_numeric.drop(columns=const_cols, inplace=True)

# Retirer les colonnes d'identifiant si elles existent
id_cols = [c for c in ["CustomerID", "CustomerTenureDays", "FirstPurchaseDaysAgo"] if c in X_numeric.columns]
if id_cols:
    X_numeric.drop(columns=id_cols, inplace=True)


def compute_vif_table(data: pd.DataFrame) -> pd.DataFrame:
    """Calcule le VIF pour chaque variable du DataFrame."""
    vif_values = []
    for i in range(data.shape[1]):
        with np.errstate(divide="ignore", invalid="ignore"):
            value = variance_inflation_factor(data.values, i)
        vif_values.append(value if np.isfinite(value) else np.inf)

    vif_df = pd.DataFrame({"Feature": data.columns, "VIF": vif_values})
    return vif_df.sort_values("VIF", ascending=False).reset_index(drop=True)


# 1) Distribution des valeurs VIF initiales par feature
initial_vif = compute_vif_table(X_numeric)
print("\n===== DISTRIBUTION INITIALE DU VIF PAR FEATURE =====")
display(initial_vif)

plt.figure(figsize=(10, max(4, 0.35 * len(initial_vif))))
sns.barplot(data=initial_vif, x="VIF", y="Feature", color="steelblue")
plt.axvline(10, color="red", linestyle="--", label="Seuil VIF = 10")
plt.title("Distribution initiale des VIF")
plt.legend()
plt.tight_layout()
plt.show()

# 2) Matrice de corrélation avant suppression itérative
corr_before_vif = X_numeric.corr()
print("\n===== MATRICE DE CORRÉLATION AVANT TRAITEMENT VIF =====")
display(corr_before_vif)

plt.figure(figsize=(12, 9))
sns.heatmap(corr_before_vif, cmap="RdBu_r", center=0, square=True)
plt.title("Matrice de corrélation avant suppression itérative")
plt.tight_layout()
plt.show()

# 3) Évolution du VIF après suppression itérative
threshold = 10
current_features = X_numeric.copy()
vif_history = []
iteration = 0

while True:
    vif_table = compute_vif_table(current_features)
    max_vif = vif_table.loc[0, "VIF"]
    worst_feature = vif_table.loc[0, "Feature"]

    vif_history.append({
        "Iteration": iteration,
        "Nb_features": current_features.shape[1],
        "Max_VIF": max_vif,
        "Worst_feature": worst_feature,
    })

    if max_vif <= threshold or current_features.shape[1] <= 1:
        final_vif_table = vif_table.copy()
        break

    current_features = current_features.drop(columns=[worst_feature])
    iteration += 1

vif_history_df = pd.DataFrame(vif_history)
print("\n===== ÉVOLUTION DU VIF APRÈS SUPPRESSION ITÉRATIVE =====")
display(vif_history_df)

plt.figure(figsize=(10, 4))
sns.lineplot(data=vif_history_df, x="Iteration", y="Max_VIF", marker="o")
plt.axhline(10, color="red", linestyle="--", label="Seuil VIF = 10")
plt.title("Évolution du VIF maximum au fil des suppressions")
plt.legend()
plt.tight_layout()
plt.show()

# 4) Variables finales retenues avec VIF < 10
final_features = current_features.columns.tolist()
print("\n===== VARIABLES FINALES RETENUES (VIF < 10) =====")
print(final_features)
display(final_vif_table)

# Optionnel: jeu de données réduit final pour la suite
X_vif_reduced = current_features.copy()
df_reduced_vif = pd.concat([X_vif_reduced, df_reduced[["Churn"]].reset_index(drop=True)], axis=1)
```

Ajout PCA — évaluation et application

Ce bloc calcule la PCA sur les variables numériques nettoyées, trace la variance expliquée cumulative, affiche la table des ratios de variance, choisit le nombre de composantes pour 95% de variance et produit un jeu réduit `df_pca`.

```python
# PCA: standardiser les numériques, évaluer variance expliquée et réduire
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Préparer les variables numériques (mêmes nettoyages que pour le VIF)
X_pca_source = df_reduced.drop(columns=["Churn"], errors="ignore").copy()
X_num = X_pca_source.select_dtypes(include=[np.number]).copy()
X_num.replace([np.inf, -np.inf], np.nan, inplace=True)
X_num.fillna(X_num.median(numeric_only=True), inplace=True)
const_cols = X_num.nunique(dropna=False)[X_num.nunique(dropna=False) <= 1].index.tolist()
if const_cols:
    X_num.drop(columns=const_cols, inplace=True)
id_cols = [c for c in ["CustomerID","CustomerTenureDays","FirstPurchaseDaysAgo"] if c in X_num.columns]
if id_cols:
    X_num.drop(columns=id_cols, inplace=True)

# Standardisation
scaler_pca = StandardScaler()
X_scaled = scaler_pca.fit_transform(X_num)

# Fit PCA complet pour évaluer
pca_full = PCA()
pca_full.fit(X_scaled)
explained = pca_full.explained_variance_ratio_
cum_explained = np.cumsum(explained)

plt.figure(figsize=(8,4))
plt.plot(np.arange(1, len(explained)+1), cum_explained, marker='o')
plt.axhline(0.95, color='red', linestyle='--', label='95% variance')
plt.xlabel('Nombre de composantes')
plt.ylabel('Variance expliquée cumulative')
plt.title('PCA — variance expliquée cumulative')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Tableau des ratios de variance
ev_df = pd.DataFrame({
    'PC': [f'PC{i+1}' for i in range(len(explained))],
    'VarianceRatio': explained,
    'CumVariance': cum_explained
})
display(ev_df.head(20))

# Choix du nombre de composantes (95% de variance expliquée)
n_components = int(np.searchsorted(cum_explained, 0.95) + 1)
print('Nombre de composantes pour 95% de variance :', n_components)

# Application de la PCA réduite
pca_reduced = PCA(n_components=n_components)
X_pca = pca_reduced.fit_transform(X_scaled)

# DataFrame résultant
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(n_components)])
df_pca = pd.concat([df_pca, df_reduced[["Churn"]].reset_index(drop=True)], axis=1)
print('Shape df_pca:', df_pca.shape)
display(df_pca.head())

# Exposer la version réduite pour la suite
X_pca_reduced = df_pca.drop(columns=["Churn"], errors='ignore')
```

 TRAITEMENT DES DONNÉES NON NUMÉRIQUES

1) Détection des colonnes catégorielles

```python
cat_cols = df.select_dtypes(include=["object", "string"]).columns.tolist()

print("\n===== COLONNES CATÉGORIELLES =====")
print(cat_cols)
print("Nombre:", len(cat_cols))

print("\n===== CARDINALITÉ =====")
display(df[cat_cols].nunique().sort_values(ascending=False))
```

2) Nettoyage des catégories

```python
def clean_cat_series(s: pd.Series) -> pd.Series:
    s = s.astype("string").str.strip()
    s = s.replace({
        "": pd.NA,
        "nan": pd.NA,
        "NaN": pd.NA,
        "None": pd.NA,
        "UNKNOWN": "Unknown",
        "unknown": "Unknown",
        "Inconnu": "Unknown",
        "INCONNU": "Unknown",
    })
    return s

for c in cat_cols:
    df[c] = clean_cat_series(df[c])

print("\n===== APRÈS NETTOYAGE =====")
display(df[cat_cols].head())
```

3) Traitement des dates

```python
if "RegistrationDate" in df.columns:
    df["RegistrationDate_parsed"] = pd.to_datetime(
        df["RegistrationDate"],
        errors="coerce",
        dayfirst=True
    )

    df["RegYear"] = df["RegistrationDate_parsed"].dt.year
    df["RegMonth"] = df["RegistrationDate_parsed"].dt.month
    df["RegDay"] = df["RegistrationDate_parsed"].dt.day
    df["RegWeekday"] = df["RegistrationDate_parsed"].dt.weekday

    print("\n===== FEATURES DATE =====")
    display(df[["RegYear","RegMonth","RegDay","RegWeekday"]].head())
```

4) Suppression des colonnes inutiles

```python
drop_cols = []

if "LastLoginIP" in df.columns:
    drop_cols.append("LastLoginIP")

if "NewsletterSubscribed" in df.columns:
    if df["NewsletterSubscribed"].nunique(dropna=False) == 1:
        drop_cols.append("NewsletterSubscribed")

if "RegistrationDate" in df.columns:
    drop_cols.append("RegistrationDate")

if "RegistrationDate_parsed" in df.columns:
    drop_cols.append("RegistrationDate_parsed")

df = df.drop(columns=list(set(drop_cols)), errors="ignore")

print("\n===== COLONNES SUPPRIMÉES =====")
print(drop_cols)

print("\n===== DATASET APRÈS SUPPRESSION =====")
display(df.head())
```

5) Ordinal Encoding

```python
ordinal_maps = {
    "AgeCategory": ["18-24","25-34","35-44","45-54","55-64","65+","Unknown"],
    "SpendingCategory": ["Low","Medium","High","VIP","Unknown"],
    "LoyaltyLevel": ["Nouveau","Jeune","Établi","Ancien","Unknown"],
    "ChurnRiskCategory": ["Faible","Moyen","Élevé","Critique","Unknown"],
    "WeekendPreference": ["Semaine","Weekend","Unknown"],
    "BasketSizeCategory": ["Petit","Moyen","Grand","Unknown"],
    "PreferredTimeOfDay": ["Matin","Midi","Après-midi","Soir","Nuit","Unknown"],
}

for col, order in ordinal_maps.items():
    if col in df.columns:
        df[col] = pd.Categorical(df[col], categories=order, ordered=True).codes
        df[col] = df[col].replace(-1, np.nan)

print("\n===== APRÈS ORDINAL ENCODING =====")
display(df.head())
```

6) One-Hot Encoding

```python
onehot_cols = []

for c in ["RFMSegment","CustomerType","FavoriteSeason","Region",
          "ProductDiversity","Gender","AccountStatus","Country"]:
    if c in df.columns:
        onehot_cols.append(c)

df = pd.get_dummies(df, columns=onehot_cols, drop_first=False, dummy_na=True)

print("\n===== APRÈS ONE-HOT ENCODING =====")
print("Shape:", df.shape)
display(df.head())
```
